```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.insert(0, str(project_root / "src"))


import cococo.layouts as layouts
```

```python
import pickle

pkl_path = "schedule_2026-04-27_14-54-03.pkl"  # change this

with open(pkl_path, "rb") as f:
    schedule = pickle.load(f)

len(schedule)
```

```python
layout_type = "triple"
m = 4
n = 4
factories = []
remove_edges = False
g, data_qubit_locs, factory_ring = layouts.gen_layout_scalable(layout_type, m, n, factories, remove_edges)
layout = {i: j for i,j in enumerate(data_qubit_locs)}
t=2
```

```python
from IPython.display import HTML
from cococo.animation_routing_html import make_clean_routing_html_animation

anim = make_clean_routing_html_animation(
    g,
    schedule,
    initial_layout=layout,
    factories=factory_ring,      # or factory_ring / factory_pos
    figsize=(18, 8),
    interval=900,
    show_active_labels=True,
    show_move_text=True,
)

HTML(anim.to_jshtml())
```
